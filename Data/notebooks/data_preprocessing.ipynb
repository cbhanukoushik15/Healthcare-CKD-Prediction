```python
# Week 1: Data Preprocessing
import pandas as pd
import numpy as np

## Load Dataset
df = pd.read_csv("kidney_disease.csv")

## Preview of Dataset
df.head()
```

```python
## Data Inspection
df.info()
```

```python
## Checking Missing Values
df.isnull().sum()
```

```python
## Data Cleaning
df.columns = df.columns.str.strip()
```

```python
# Convert to string first
df['classification'] = df['classification'].astype(str)

# Remove spaces
df['classification'] = df['classification'].str.strip()

# Convert to numbers
df['classification'] = df['classification'].map({'ckd':1, 'notckd':0})
```

```python
df['classification'].value_counts()
```

```python
df.replace('?', np.nan, inplace=True)
```

```python
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='ignore')
```

```python
# Numerical columns
for col in df.select_dtypes(include=['float64', 'int64']):
    df[col].fillna(df[col].median(), inplace=True)

# Categorical columns
for col in df.select_dtypes(include=['object']):
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
## Encoding Categorical Variables
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in df.select_dtypes(include=['object']):
    df[col] = le.fit_transform(df[col])
```

```python
## Final Data Check
df.isnull().sum()
```

```python
df.rename(columns={
    'bp': 'blood_pressure',
    'sg': 'specific_gravity',
    'al': 'albumin',
    'su': 'sugar',
    'rbc': 'red_blood_cells',
    'pc': 'pus_cell',
    'pcc': 'pus_cell_clumps',
    'ba': 'bacteria',
    'bgr': 'blood_glucose_random',
    'bu': 'blood_urea',
    'sc': 'serum_creatinine',
    'sod': 'sodium',
    'pot': 'potassium',
    'hemo': 'hemoglobin',
    'pcv': 'packed_cell_volume',
    'wc': 'white_blood_cell_count',
    'rc': 'red_blood_cell_count',
    'htn': 'hypertension',
    'dm': 'diabetes_mellitus',
    'cad': 'coronary_artery_disease',
    'appet': 'appetite',
    'pe': 'pedal_edema',
    'ane': 'anemia'
}, inplace=True)
```

```python
## Saving Cleaned Dataset
df.to_csv("cleaned_ckd.csv", index=False)
```

## Clinical Rationale for Data Preprocessing

- Median was used for numerical features because medical datasets often contain outliers.
- Mode was used for categorical variables to preserve the most common clinical condition.
- This ensures realistic and unbiased data representation.
