# Week 3: Predictive Modeling and Algorithm Selection

In this phase, machine learning models are developed to predict the presence of Chronic Kidney Disease (CKD) using the processed dataset.

The dataset is divided into training and testing sets, and multiple classification models are trained and evaluated.

The objective is to identify the most effective model for predicting CKD, with a strong focus on recall to minimize false negatives, which is critical in healthcare applications.

```python
## Importing Libraries
import pandas as pd

## Loading Dataset
df = pd.read_csv("cleaned_ckd.csv")

print(df.shape)

## Preview of Dataset
df.head()
```

```python
## Data Inspection
df.info()
```

```python
## feature Selection and Target Variable
X = df.drop('classification', axis=1)
y = df['classification']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=10, max_depth=3)
rf.fit(X_train, y_train)

print("Model trained successfully")
```

```python
## Train-Test Split
from sklearn.metrics import confusion_matrix, classification_report

y_pred = rf.predict(X_test)

print("Confusion Matrix:\n")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
```

## Model Training

Two models are trained:
- Logistic Regression
- Random Forest

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=100,
    max_depth=8,
    min_samples_split=5,
    random_state=42
)

rf.fit(X_train, y_train)
```

## Model Evaluation

Recall is the most important metric in this project because false negatives (failing to identify a CKD patient) can have serious medical consequences.

A high recall ensures that most CKD patients are correctly identified by the model.

```python
y_pred_lr = lr.predict(X_test)

print("Logistic Regression:\n")
print(confusion_matrix(y_test, y_pred_lr))
print(classification_report(y_test, y_pred_lr))
```

```python
y_pred_tuned = rf.predict(X_test)
print("Tuned Random Forest:\n")
print(confusion_matrix(y_test, y_pred_tuned))
print(classification_report(y_test, y_pred_tuned))
```

```python
from sklearn.metrics import roc_auc_score

y_prob = rf.predict_proba(X_test)[:,1]
roc_score = roc_auc_score(y_test, y_prob)

print("ROC-AUC Score:", roc_score)
```

The ROC-AUC score measures the model’s ability to distinguish between CKD and non-CKD patients.

A higher ROC-AUC score indicates better model performance.

## Model Comparison

Random Forest performs better than Logistic Regression in capturing complex relationships within the data.

It achieves higher recall, making it more suitable for healthcare prediction where correctly identifying CKD patients is critical.

Therefore, Random Forest is selected as the final model for this project.

## Hyperparameter Tuning

The Random Forest model was tuned by adjusting parameters such as the number of trees and maximum depth.

This improved the model's performance and helped achieve better recall, making it more effective for identifying CKD patients.

## Conclusion

The predictive models were successfully developed and evaluated.

Random Forest was selected as the best-performing model due to its higher recall and ability to identify CKD patients effectively.

These results demonstrate the potential of machine learning in assisting early detection of chronic diseases.

The model demonstrates strong performance and can be used to support early detection of CKD in healthcare applications.
