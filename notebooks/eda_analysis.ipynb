This section explores the dataset to understand patterns, relationships, and differences between CKD and Non-CKD patients.

Target Distribution

Feature Distributions

CKD vs Non-CKD Comparison

Correlation Analysis

Summary of Insights

```python
## Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Load Dataset
df = pd.read_csv("cleaned_ckd.csv")
```

```python
## Preview of Dataset
df.head()
```

## Target Distribution Analysis



```python
df['classification'].value_counts().plot(kind='bar')
plt.title("CKD vs Non-CKD Distribution")
plt.xlabel("Class (1 = CKD, 0 = Not CKD)")
plt.ylabel("Count")
plt.show()
```

## Target Distribution Insight

The dataset shows that CKD patients are more frequent than non-CKD patients (approximately 250 vs 150).

This indicates a slight class imbalance, which is important to consider during model building.

## Age Distribution Analysis



```python
df['age'].hist()
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

## Age Distribution Insight

The distribution shows that older patients are more common in the dataset, suggesting that CKD risk may increase with age.

## Blood Pressure Analysis



```python
df['blood_pressure'].hist()
plt.title("Blood Pressure Distribution")
plt.xlabel("Blood Pressure")
plt.ylabel("Frequency")
plt.show()
```

## Blood Pressure Insight

Blood pressure shows noticeable differences between CKD and non-CKD patients.

CKD patients tend to have higher blood pressure levels, indicating that hypertension is a significant risk factor associated with kidney disease.

## Serum Creatinine Analysis



```python
df['serum_creatinine'].hist()
plt.title("Serum Creatinine Distribution")
plt.show()
```

## Serum Creatinine Insight

CKD patients exhibit significantly higher serum creatinine levels, indicating impaired kidney function.

The presence of extreme values suggests severe kidney damage in some patients.

## Hemoglobin Analysis

```python

df['hemoglobin'].hist()
plt.title("Hemoglobin Distribution")
plt.show()
```

## Hemoglobin Insight

CKD patients have lower hemoglobin levels.

This indicates anemia is common in CKD.

## Serum Creatinine Analysis

```python
sns.boxplot(x='classification', y='serum_creatinine', data=df)
plt.title("Serum Creatinine vs CKD")
plt.xlabel("CKD (1) vs Non-CKD (0)")
plt.ylabel("Serum Creatinine")
plt.show()
```

## Serum Creatinine Insight

CKD patients have significantly higher serum creatinine levels compared to non-CKD patients.

There is a wider distribution and presence of extreme values (outliers) in CKD patients, indicating severe kidney dysfunction in some cases.

This confirms that serum creatinine is a strong biomarker for Chronic Kidney Disease.

## Hemoglobin vs CKD Analysis

```python
sns.boxplot(x='classification', y='hemoglobin', data=df)
plt.title("Hemoglobin vs CKD")
plt.show()
```

## Hemoglobin vs CKD Insight

CKD patients tend to have lower hemoglobin levels, which may indicate anemia, a common condition in kidney disease.

## Blood Pressure vs CKD Analysis

```python
sns.boxplot(x='classification', y='blood_pressure', data=df)
plt.title("Blood Pressure vs CKD")
plt.show()
```

## Blood Pressure vs CKD Insight

Blood pressure shows noticeable differences between CKD and non-CKD patients

## Correlation Heatmap Analysis

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

## Correlation Heatmap Analysis

The heatmap shows relationships between different medical features.

Serum creatinine and blood urea have strong positive correlation with CKD, indicating their importance in identifying kidney disease.

Hemoglobin shows a negative relationship with CKD, suggesting lower levels in CKD patients.

Some features are also highly correlated with each other, indicating possible multicollinearity.

This analysis helps in identifying important features and avoiding multicollinearity during model building.

## Key Feature Relationships

- Serum creatinine is strongly associated with CKD
- Blood urea levels increase in CKD patients
- Hemoglobin shows negative correlation with CKD (lower in CKD patients)

## Summary of Insights

- CKD patients have higher serum creatinine levels
- Hemoglobin levels are lower in CKD patients
- Blood pressure shows variation across patients
- Strong relationships exist between key medical features

These insights help in identifying important features for predictive modeling.

These insights will be used in the next phase for building predictive machine learning models.
